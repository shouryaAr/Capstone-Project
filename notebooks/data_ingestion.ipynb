```python
import pandas as pd
from pathlib import Path
file_path = Path.cwd().parent / 'data' / 'raw' 
df = pd.read_csv(file_path / '01_fund_master.csv')
print("Columns in Fund Master:")
print(list(df.columns))
print(df.head(2))
```

```python
print("Unique amfi codes in Fund Master:")
print(df['amfi_code'].unique())
print("Unique fund houses in Fund Master:")
print(list(df['fund_house'].unique()))
print("Unique categories in Fund Master:")
print(list(df['category'].unique()))
print("No. of unique sub-categories in Fund Master:")
print(df['sub_category'].nunique())
print("Unique risk grades in Fund Master:")
print(list(df['risk_category'].unique()))
```

```python
nav_df = pd.read_csv(file_path / '02_nav_history.csv')
nav_df.columns
fund_master_codes = set(df['amfi_code'])
nav_codes = set(nav_df['amfi_code'])
missing_codes = fund_master_codes - nav_codes
if missing_codes == 0:
    print(f"Missing NAV data for {len(missing_codes)} codes")
else:
    print("No missing NAV data for any code")
```
